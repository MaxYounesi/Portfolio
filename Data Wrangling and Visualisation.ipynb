#### Investigate all features for missing values

```python
# Check the overall number of missing values for all variables
auto.isnull().sum()
```

With regards to the categorical features, the mode of the variable (the most frequent value) is used 
to override the missing value. And for the numeric features, the median of the variables is utilised.

```python
# Find the mode for 'Num-of-doors'
num_doors_mode = auto['Num-of-doors'].mode()
num_doors_mode
```

```python
# Replace NULL values for 'Num-of-doors' with the mode ('FOUR')
auto['Num-of-doors'].fillna('FOUR', inplace=True)
```

```python
# Inspect the results
auto['Num-of-doors'].value_counts()
```

```python
# Replacing NULL values for numeric variables with the median value
auto['Normalised-losses'].fillna(auto['Normalised-losses'].median(), inplace=True)
auto['Bore'].fillna(auto['Bore'].median(), inplace=True)
auto['Stroke'].fillna(auto['Stroke'].median(), inplace=True)
auto['Horsepower'].fillna(auto['Horsepower'].median(), inplace=True)
auto['Peak-rpm'].fillna(auto['Peak-rpm'].median(), inplace=True)
auto['Price'].fillna(auto['Price'].median(), inplace=True)
```

```python
# Inspect the results by checking the overall number of missing values for all variables
auto.isnull().sum()
```

#### Data type conversion

```python
# Convert data types from 'object' to 'nominal' and 'ordinal'
auto['Symboling'] = pd.Categorical(auto['Symboling'],ordered=True,
                                  categories=[-3, -2, -1, 0, 1, 2, 3])
auto['Make'] = pd.Categorical(auto['Make'])
auto['Fuel-type'] = pd.Categorical(auto['Fuel-type'])
auto['Aspiration'] = pd.Categorical(auto['Aspiration'])
auto['Num-of-doors'] = pd.Categorical(auto['Num-of-doors'],
                                      ordered=True,
                                     categories=['TWO','FOUR'])
auto['Body-style'] = pd.Categorical(auto['Body-style'])
auto['Drive-wheels'] = pd.Categorical(auto['Drive-wheels'])
auto['Engine-location'] = pd.Categorical(auto['Engine-location'])
auto['Engine-type'] = pd.Categorical(auto['Engine-type'])
auto['Num-of-cylinders'] = pd.Categorical(auto['Num-of-cylinders'],
                                          ordered=True,
                                         categories=['TWO','THREE','FOUR','FIVE','SIX',
                                                    'EIGHT','TWELVE'])
auto['Fuel-system'] = pd.Categorical(auto['Fuel-system'])
```

#### 1. Univariate Charts

Choose 1 column with nominal values, 1 column with ordinal Values, and 1 column with numerical value

```python
# Install and load seaborn
!pip install seaborn
import seaborn as sns
```

```python
# Nominal variable = 'Body-style'
# Chart type = Bar graph
import matplotlib.pyplot as plt
auto['Body-style'].value_counts().plot(kind = 'bar')
plt.xlabel('Body-style')
plt.ylabel('Count of Automobiles')
plt.legend(loc="upper right")
plt.title('Distribution of Automobiles per Body-style')
```

```python
# Ordinal variable = 'Num-of-cylinders'
# Chart type = Bar graph
num_cylinders = auto['Num-of-cylinders'].value_counts().reset_index()
print(num_cylinders)
num_cylinders['index'] = pd.Categorical(num_cylinders['index'],
                                        ordered = True,
                                        categories = ['TWO', 'THREE', 'FOUR', 'FIVE', 'SIX', 'EIGHT', 'TWELVE'])
num_cylinders.sort_values(by = ['index'], inplace=True)
num_cylinders.plot.bar(x='index', y='Num-of-cylinders')
plt.xlabel('Number of Cylinders')
plt.ylabel('Count of Automobiles')
plt.title('Distribution of Automobiles per Number of Cylinders')
```

```python
# Numeric variable = 'Price'
# Chart type = Histogram
auto['Price'].plot(kind='hist', bins=30)
plt.title('Automobiles - Price Info')
plt.xlabel('Price')
plt.grid()
plt.show()
```

```python
# Numeric variable = 'Price'
# Chart type = Density chart
auto['Price'].plot(kind='density')
plt.title('Automobiles - Price Info')
plt.xlabel('Price')
plt.grid()
plt.show()
```
